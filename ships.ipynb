```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def read_html(link):
  response = requests.get(link)
  return(BeautifulSoup(response.text, 'html'))

def get_texts(link_nodes):
  return([x.string for x in link_nodes])

def clean_lower(my_texts):
  return([x.lower().strip() for x in my_texts])
```

```python
# step 1: get the links

t = read_html('https://www.boat24.com/en/secondhandboats/')
links_node = t.find_all(class_='blurb__title')
links = [x.find('a')['href'] for x in links_node]


```

```python
def get_one_ship(link):
  t = read_html(link)
  values = get_texts(t.find_all(class_='list__value'))
  keys = get_texts(t.find_all(class_='list__key'))
  data = dict(zip(keys, values))
  return(data)
```

```python
data_list = list(map(get_one_ship, links[1:5]))
df = pd.DataFrame(data_list)
df
```
