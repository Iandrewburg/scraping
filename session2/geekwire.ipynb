```python
# Step 1: Create a df of the first page, 5 values, use read_cluoud becosu read_html ( the simple get request is blocked by the site)
t = read_cloud('https://www.geekwire.com/geeklife/')
title = get_texts(t.select('.entry-title a'))
links = get_links(t.select('.entry-title a'))
authors = get_texts(t.select('.fn'))
dates = get_texts(t.select('.published'))
summary = strip_texts(get_texts(t.select('.entry-summary')))
pd.DataFrame({'title' :  title, 'link' : links, 'summary' : summary, 'author': authors, 'date': dates  })
```

```python
# Step 2: Put the code into a function, one link is the input, df the output
def read_one_page(link):
    print(link)
    t = read_cloud(link)
    title = get_texts(t.select('.entry-title a'))
    links = get_links(t.select('.entry-title a'))
    authors = get_texts(t.select('.fn'))
    dates = get_texts(t.select('.published'))
    summary = strip_texts(get_texts(t.select('.entry-summary')))
    df = pd.DataFrame({'title' :  title, 'link' : links, 'summary' : summary, 'author': authors, 'date': dates  })
    return(df)
```

```python
# test it
one_page = read_one_page('https://www.geekwire.com/geeklife/')
one_page
```

```python
# Step 4: map your function to your list
list_of_dfs = list(map(read_one_page, all_links))
```

```python
# Update the function
def read_one_page(link):
    print(link)
    t = read_cloud(link)
    title = get_texts(t.select('.entry-title a'))
    links = get_links(t.select('.entry-title a'))
    authors = get_texts(t.select('.text-uppercase strong'))
    authors = [x.replace('by ', '') for x in authors]
    dates = get_texts(t.select('.published'))
    summary = strip_texts(get_texts(t.select('.entry-summary')))
    df = pd.DataFrame({'title' :  title, 'link' : links, 'summary' : summary, 'author': authors, 'date': dates  })
    return(df)
```

```python
# Step 4 again: map your function to your list
list_of_dfs = list(map(read_one_page, all_links))
```
