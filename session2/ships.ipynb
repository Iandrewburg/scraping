```python
from scrapethat import *
# step 1: get the links of every single ship on the first main page
t = read_html('https://www.boat24.com/en/secondhandboats/')
links_node = t.find_all(class_='blurb__title')
links = [x.find('a')['href'] for x in links_node]
print(len(links))
links
```

```python
# step 2: get the information of a single ship
t = read_html('https://www.boat24.com/en/powerboats/malibu/malibu-response-txi/detail/560962/')
values = get_texts(t.find_all(class_='list__value'))
keys = get_texts(t.find_all(class_='list__key'))
data = dict(zip(keys, values))
data
```

```python
# step 3: Put it into a function. function get one link and return with a dictionary
def get_one_ship(link):
  t = read_html(link)
  values = get_texts(t.find_all(class_='list__value'))
  keys = get_texts(t.find_all(class_='list__key'))
  data = dict(zip(keys, values))
  return(data)

get_one_ship('https://www.boat24.com/en/powerboats/striker/striker-45f/detail/540744/')
```

```python
# step 4: map your function to your link list (here i just selecting 4 link)
data_list = list(map(get_one_ship, links[1:5]))
df = pd.DataFrame(data_list)
df
```

```python
# step 5: go bact to step 1 and create more link, so you can download more data.
# the links are changing like that:
# https://www.boat24.com/en/secondhandboats/
# https://www.boat24.com/en/secondhandboats/?page=20
# https://www.boat24.com/en/secondhandboats/?page=40

main_pages = ['https://www.boat24.com/en/secondhandboats/']
# generate the first 10 page
print(list(range(20, 100, 20)))
main_pages.extend([f"https://www.boat24.com/en/secondhandboats/?page={x}" for x in range(20, 100, 20)])
main_pages
```

```python
# step 6: get the boat links from the main_pages links

all_links = []

for main_page in main_pages:
    t = read_html(main_page)
    links_node = t.find_all(class_='blurb__title')
    newlinks = [x.find('a')['href'] for x in links_node]
    all_links.extend(newlinks)
print(len(all_links))
all_links
```

```python
# step 6 in nicer way create a function to get the links than apply to main_pages_link
def get_links(link):
    t = read_html(link)
    links_node = t.find_all(class_='blurb__title')
    return([x.find('a')['href'] for x in links_node])

list_of_links = list(map(get_links, main_pages))
```

```python
# you can see it is a list of list
list_of_links
```

```python
# to flatten that list you have to use that code
from itertools import chain
flattened_list = list(chain.from_iterable(list_of_links))
```
