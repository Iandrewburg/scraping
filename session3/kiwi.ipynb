# Locations

```python
import requests

headers = {
    'accept': 'application/json',
    'apikey': '',
}

params = {
    'term': 'VIE',
    'locale': 'en-US',
    'location_types': 'airport',
    'limit': '1000',
    'active_only': 'true',
}

response = requests.get('https://api.tequila.kiwi.com/locations/query', params=params, headers=headers)

```

```python
data = response.json()
data.keys()
```

```python
from pandas import json_normalize

df = json_normalize(response.json()['locations'])
```
